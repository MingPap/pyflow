```python
# Demo notebook
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np
from PIL import Image
import time
import argparse
import os
import pyflow
```

```python
# Load input files
examples_dir = "./examples"
img1 = Image.open(os.path.join(examples_dir, 'car1.jpg'))
img2 = Image.open(os.path.join(examples_dir, 'car2.jpg'))
```

```python
# Build arrays
im1 = np.array(img1).astype(float) / 255.
im2 = np.array(img2).astype(float) / 255.
```

```python
# Flow Options:
alpha = 0.012 # default 0.012
ratio = 0.75 # default 0.75
minWidth = 20 # default 20
nOuterFPIterations = 7 # default 7
nInnerFPIterations = 1 # default 1
nSORIterations = 30 # default 30
colType = 0  # 0 or default:RGB, 1:GRAY (but pass gray image with shape (h,w,1))
threshold = 0.000005
```

```python
# Compute flow
s = time.time()
u, v, im2W = pyflow.coarse2fine_flow(
    im1, im2, alpha, ratio, minWidth, nOuterFPIterations, nInnerFPIterations,
    nSORIterations, colType, verbose = True, threshold = threshold)
e = time.time()
print('Time Taken: %.2f seconds for image of size (%d, %d, %d)' % (e - s, im1.shape[0], im1.shape[1], im1.shape[2]))
```

```python
Image.fromarray(np.uint8(im2W*255))
```

```python
def create_diagonal_gradient_image(height, width, depth):
  """Creates pyramid image. Useful for testing.

  For example, pyramid_image(5, 6, 1) looks like:
  # [[[ 5.  4.  3.  2.  1.  0.]
  #   [ 6.  5.  4.  3.  2.  1.]
  #   [ 7.  6.  5.  4.  3.  2.]
  #   [ 8.  7.  6.  5.  4.  3.]
  #   [ 9.  8.  7.  6.  5.  4.]]]

  Args:
    height: height of image
    width: width of image
    depth: depth of image

  Returns:
    pyramid image
  """
  row = np.arange(height)
  col = np.arange(width)[::-1]
  image_layer = np.expand_dims(row, 1) + col
  image_layer = np.expand_dims(image_layer, 2)

  image = image_layer
  for i in range(1, depth):
    image = np.concatenate((image, image_layer * pow(10, i)), 2)

  return image.astype(np.float64) / np.max(image)
```

```python
gradient = create_diagonal_gradient_image(im1.shape[0], im1.shape[1], 1)
```

```python
gradient.shape, np.min(gradient), np.max(gradient), np.mean(gradient), np.std(gradient)
```

```python
Image.fromarray(np.uint8(gradient[:,:,0] * 255.))
```

```python
# Warp greyscale image based on optical flow
s = time.time()
gradientW = pyflow.warp_mask_flow(gradient, u, v)
e = time.time()
print('Time Taken: %.2f seconds for image of size (%d, %d, %d)' % (e - s, gradient.shape[0], gradient.shape[1], gradient.shape[2]))
```

```python
gradientW.shape, np.min(gradientW), np.max(gradientW), np.mean(gradientW), np.std(gradientW), gradientW.dtype
```

```python
Image.fromarray(np.uint8(gradientW[:,:,0] * 255.))
```
